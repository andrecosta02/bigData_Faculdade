```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("ExemploLeitura").getOrCreate()

# Leitura de um arquivo CSV
df_pyspark = spark.read.csv("/root/.cache/kagglehub/datasets/bwandowando/philippine-cities-weather-data-2000-2009/versions/1/daily_data_combined.csv", header=True, inferSchema=True)
```

```python
#Nomes das colunas
for i in df_pandas_weather.columns:
  print(i)
```

```python
# Seleciona colunas específicas - Pandas
df_selecionado_pandas = df_pandas_weather[['city_name', 'latitude', 'longitude', 'weather_code','temperature_2m_max', 'temperature_2m_min']]
print(df_selecionado_pandas)
```

```python
# Seleciona colunas específicas - PySpark
df_selecionado_spark = df_pyspark.select('city_name', 'latitude', 'longitude', 'weather_code','temperature_2m_max', 'temperature_2m_min')
```

```python
# Filtra linhas onde coluna1 = 'Alaminos' - Pandas
df_filtrado_pandas = df_pandas_weather[df_pandas_weather['city_name'] == 'Alaminos']
```

```python
# Filtra linhas onde coluna1 = 'Alaminos'- PySpark
df_filtrado_spark = df_pyspark.filter(df_pyspark.city_name == 'Alaminos')
```

```python
# Calcula a média de coluna1 agrupada por coluna7 - Pandas
df_agregado_pandas = df_pandas_weather.groupby('city_name')['temperature_2m_max'].mean().reset_index()
print(df_agregado_pandas)
```

```python
# Calcula a média de coluna1 agrupada por coluna7 - PySpark
from pyspark.sql.functions import avg

# Calcula a média de coluna1 agrupada por coluna2
df_agregado_spark = df_pyspark.groupBy("city_name").agg(avg("temperature_2m_max").alias("media_days_left"))
df_agregado_spark.show(5)
```
